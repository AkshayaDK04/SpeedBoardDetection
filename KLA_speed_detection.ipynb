```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import easyocr
import csv
import os

# Load the YOLO model
model_path = "/content/gdrive/MyDrive/last.pt"
model = YOLO(model_path)

# Perform inference on the image

csv_file_path = "/content/gdrive/MyDrive/speeds.csv"

dataset_path='/content/gdrive/MyDrive/DataSet_val/DataSet/Milestone 2'

# Read the original image



with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Image", "Speed"])  # Writing the header

    # Iterate over the images in the dataset
    for imgs in os.listdir(dataset_path):
        img_path = os.path.join(dataset_path, imgs)
        img_path = os.path.normpath(img_path)
        results = model(img_path)
        image = cv2.imread(img_path)
        image_with_detections = results[0].plot()

        for i, result in enumerate(results[0].boxes):
            x1, y1, x2, y2 = map(int, result.xyxy[0])
            cropped_image = image[y1:y2, x1:x2]
            cropped_image_path = f"cropped_detected_{i}.jpg"
            cv2.imwrite(cropped_image_path, cropped_image)

            reader = easyocr.Reader(['en'])
            r_text = reader.readtext(cropped_image_path)

            # Extract text and save it
            for (bbox, text, prob) in r_text:
              if(text.isdigit()):
                print(text)
                writer.writerow([imgs, text])
        plt.imshow(cv2.cvtColor(image_with_detections, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
```

```python
import cv2
from ultralytics import YOLO
import easyocr

# Load the YOLO model
model = YOLO('yolov5s.pt')  # Use a suitable YOLO model

# Load the image
image_path = "/content/gdrive/MyDrive/DataSet_val/DataSet/Milestone 3/Picture4.jpg"
image = cv2.imread(image_path)

# Perform object detection
results = model(image)
image_with_detections = results[0].plot()

# Extract bounding boxes
for detection in results[0].boxes:
    x1, y1, x2, y2 = map(int, detection.xyxy[0])
    roi = image[y1:y2, x1:x2]

    # Further process each ROI...
    # Convert the ROI to grayscale
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)

# Apply thresholding to binarize the image
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)
    reader = easyocr.Reader(['en'])

# Read the text from the processed image
    text_results = reader.readtext(thresh)

# Print the detected text
    for (bbox, text, prob) in text_results:
        print(f"Detected text: {text} with confidence: {prob}")
plt.imshow(cv2.cvtColor(image_with_detections, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
